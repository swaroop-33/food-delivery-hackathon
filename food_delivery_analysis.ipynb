```python
import pandas as pd

df = pd.read_csv("final_food_delivery_dataset.csv")
df.head()
```

```python
df['order_date'] = pd.to_datetime(df['order_date'], dayfirst=True)

df.groupby(df['order_date'].dt.to_period('M'))['total_amount'].sum()
```

```python
df.groupby('user_id')['order_id'].count().sort_values(ascending=False).head()
```

```python
df.groupby('city')['total_amount'].sum().sort_values(ascending=False)
```

```python
df.groupby('cuisine')['total_amount'].sum().sort_values(ascending=False)
```

```python
df.groupby('membership')['total_amount'].mean()
```

```python
df['month'] = df['order_date'].dt.month
df.groupby('month')['total_amount'].sum()
```
